```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# criando o grafo de exemplo
G = nx.Graph()
G.add_edges_from([(1,2),(1,3),(1,4)])
nx.draw_networkx(G)
```

# Medidas em grafos

## Cálculo do **Grau** de um nó (O que é grau?)

```python
G.degree[1]
```

## Cálculo do **Diâmetro** de uma rede (O que é diâmetro ?)

```python
# Cálculo do diâmetro
diameter = nx.diameter(G)
print("Diâmetro da rede: ",diameter)
```
