##Import of Libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
```

##Directed Graph Methods

```python
DG = nx.DiGraph()
DG.add_weighted_edges_from([(1,2,0.8),(3,1,0.4),(1,5,0.9)])
DG.in_degree(1,weight='weight')
```

```python
nx.draw_networkx(DG)
```

```python
DG.degree(1,weight='weight')
```

```python
list(DG.successors(1))
```

```python
list(DG.predecessors(1))
```

```python
nx.draw_networkx(DG)
```

We can create an undirected graph *H* from a directed graph *DG*.

```python
H = nx.Graph(DG)
nx.draw_networkx(H)
```
