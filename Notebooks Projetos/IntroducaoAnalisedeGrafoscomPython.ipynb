# Grafos

Um grafo √© uma estrutura matem√°tica que consiste em um conjunto de v√©rtices (tamb√©m chamados de n√≥s) e um conjunto de arestas que conectam pares de v√©rtices.

A seguir temos um grafo com 6 n√≥s (c√≠rculos amarelos) e 7 v√©rtices (linhas pretas).

<p align="center"><img src="https://github.com/paulosrl/imagens/blob/master/grafo.png?raw=true"></p>

Os grafos permitem codificar relacionamentos entre pares de objetos. Os objetos s√£o os v√©rtices do grafo e os relacionamentos s√£o suas arestas.

Os grafos s√£o amplamente utilizados na teoria dos grafos, um ramo da matem√°tica discreta, e t√™m diversas aplica√ß√µes em ci√™ncia da computa√ß√£o, redes, log√≠stica, transporte, entre outros campos.

Os grafos s√£o fundamentais em algoritmos, pois muitos problemas podem ser modelados e resolvidos de forma eficiente usando t√©cnicas baseadas em grafos. Eles s√£o uma ferramenta poderosa para representar e analisar rela√ß√µes e estruturas complexas em diversos contextos.

## Principais conceitos

Existem diferentes tipos de grafos, e suas propriedades variam com base na natureza e nas rela√ß√µes das arestas. Alguns conceitos-chave em teoria dos grafos incluem:

### N√≥s e v√©rtices

Representam entidades, sejam elas pessoas, perfis do Twitter, empresas, p√°ginas ou usu√°rios do Facebook, canais ou v√≠deos do Youtube, etc. Estes n√≥s podem possuir atributos ou caracter√≠sticas espec√≠ficas, tais como tipos, tamanho e peso.



### Arestas ou v√©rtices

S√£o as conex√µes entre os n√≥s, representando as liga√ß√µes entre as entidades.

Em uma rede social, por exemplo, estas rela√ß√µes podem representar intera√ß√µes como curtidas, retweets (no caso do Twitter/X) ou conex√µes entre canais de YouTube.



## Tipos de grafo

### Grafo n√£o direcionado

Um grafo no qual as arestas n√£o t√™m dire√ß√£o, representando uma rela√ß√£o bidirecional entre os v√©rtices.

Um exemplo de grafo n√£o direcionado seria uma rela√ß√£o de amizade em uma rede social, o v√≠nculo (aresta) estabelece a rela√ß√£o entre duas pessoas sem usar uma dire√ß√£o.

No exemplo a seguir, Carlos √© amigo de Maria e Jo√£o.

### Grafo direcionado

Um grafo no qual as arestas t√™m uma dire√ß√£o espec√≠fica, indicando a rela√ß√£o de um v√©rtice de origem para um v√©rtice de destino.

A liga√ß√£o entre os n√≥s n√£o √© necessariamente m√∫tua, por exemplo:

No Instagram, por exemplo, √© poss√≠vel seguir algu√©m sem que a outra pessoa siga de volta.

No exemplo a seguir Jo√£o e Maria seguem Carlos, mas Carlos n√£o segue Jo√£o e Maria.

# Criando grafos direcionados


```python
# cria um grafo direcionado, o sentido da seta √© sempre do n√≥ da esquerda para direita
G = nx.DiGraph([(0,3),(1,3),(2,4),(3,6),(5,6)])
nx.draw_networkx(G)
```

```python
# cria grupo de n√≥s por colunas
nos_esquerda = [0,1,2]
nos_meio = [3,4]
nos_direita = [5,6]
# ajustando a posi√ß√£o dos n√≥s de acordo com as colunas
# posiciona os n√≥s de acordo com o grupo
# n√≥s a esquerda
pos = {n: (0,i) for i,n in enumerate(nos_esquerda)}
# n√≥s do meio
pos.update({n: (1,i+0.5) for i,n in enumerate(nos_meio)})
# n√≥s a direita
pos.update({n: (2,i+0.5) for i,n in enumerate(nos_direita)})
# desenha a rede com o atributo pos
nx.draw_networkx(G,pos)
```

## Formato bipartido

Um grafo bipartido √© um tipo de grafo que tem seus v√©rtices divididos em dois conjuntos disjuntos, de tal forma que n√£o h√° arestas entre v√©rtices do mesmo conjunto.

```python
G = nx.bipartite.gnmk_random_graph(3, 5, 10, seed=123)
top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G, top)
nx.draw_networkx(G,pos)
```

## Formato multipartido

```python
# Um grafo multipartido √© um tipo de grafo que tem seus v√©rtices
# divididos em v√°rios conjuntos disjuntos, de tal forma que n√£o
# h√° arestas entre v√©rtices do mesmo conjunto.
G = nx.complete_multipartite_graph(10, 5, 3)
pos = nx.multipartite_layout(G)
nx.draw_networkx(G,pos)
```
