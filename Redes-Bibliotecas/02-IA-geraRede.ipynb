```python
# Ete código gerado por IA. Foram feitas algumas adaptações para a geração do grafo.
# prompt utilizado: 
# gere um grafo não direcionado utilizando a biblioteca networkx python, 
# leia os nós e as arestas dos arquivos "nos.csv" e "arestas.csv"

# Importar a biblioteca pandas para manipular arquivos csv
import pandas as pd

# Importar a biblioteca networkx para criar e visualizar grafos
import networkx as nx

# Importar a biblioteca matplotlib para plotar o grafo
import matplotlib.pyplot as plt

# Ler o arquivo csv que contém os nós do grafo
# O arquivo deve ter uma coluna chamada 'nó' com os nomes dos nós
# O arquivo deve estar na mesma pasta que o código python
nos = pd.read_csv('nos.csv')

# Ler o arquivo csv que contém as arestas do grafo
# O arquivo deve ter duas colunas chamadas 'origem' e 'destino' com os nomes dos nós conectados por cada aresta
# O arquivo deve estar na mesma pasta que o código python
arestas = pd.read_csv('arestas.csv')

# Criar um grafo não direcionado vazio usando networkx
G = nx.Graph()

# Adicionar os nós do grafo usando a coluna 'nó' do dataframe nós
G.add_nodes_from(nos['no'])

# Adicionar as arestas do grafo usando as colunas 'origem' e 'destino' do dataframe arestas
G.add_edges_from(zip(arestas['origem'], arestas['destino']))

# Plotar o grafo usando matplotlib
plt.figure(figsize=(8, 6)) # Ajustar o tamanho da figura
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='black') # Desenhar o grafo com rótulos, cores e estilo
plt.show() # Mostrar a figura
```
