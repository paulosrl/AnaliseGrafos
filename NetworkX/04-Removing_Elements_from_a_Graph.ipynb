##Import of Libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
```

##Removing Elements

```python
G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(1,4)
G.add_nodes_from("paper")
G.add_edge(3,'p')
nx.draw_networkx(G)
```

```python
G.remove_node(2)
```

```python
nx.draw_networkx(G)
```

```python
G.remove_edge(1,4)
```

```python
nx.draw_networkx(G)
```

```python
G.remove_nodes_from(['e',4])
```

```python
nx.draw_networkx(G)
```

```python
G.remove_edges_from([('p',3),(1,3)])
```

```python
nx.draw_networkx(G)
```
