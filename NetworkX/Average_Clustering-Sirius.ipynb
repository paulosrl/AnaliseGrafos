##Import of Libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Average Clustering

```python
# criaÃ§Ã£o do grafo
G = nx.Graph()

G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (3, 5),  (4,5),(4,2),(1,4),(5,1),(5,2)])

nx.draw(G,with_labels=True)

# cÃ¡lculo da mÃ©dia do coeficiente de clustering
# quando todos os nÃ³s estÃ£o ligados entre si, o valor serÃ¡ igual a 1
# quanto menor o coeficiente de clustering, menos conexÃµes existirÃ¡ entre os nÃ³s.

avg_clustering = nx.average_clustering(G)

print("MÃ©dia do Coeficiente de clustering: ",avg_clustering)
```
