##Import of Libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Connectivity

```python
# Create a graph
G = nx.Graph()
G.add_edges_from([(1,2),(1,3),(2,3),(3,4)])

nx.draw(G,with_labels=True,node_color='lightblue',edge_color='gray')
```

```python
# Compute the node connectivity for all pairs of nodes
node_connectivity = nx.all_pairs_node_connectivity(G)

# Print the node connectivity for each pair of nodes

for nodes,connectivity in node_connectivity.items():
  print(f"Node connectivity between nodes {nodes} : {connectivity}")
```
