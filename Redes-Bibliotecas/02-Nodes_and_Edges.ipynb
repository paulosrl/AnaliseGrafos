##Import of Libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
```

##Nodes and Edges

###Nodes
The graph G can be grown in several ways.

To get started though we’ll look at simple manipulations. We can add one node at a time

```python
# adição de nós
G = nx.Graph()
G.add_node(1)
# add nodes from any iterable container, such as a list
G.add_nodes_from([2,3])
nx.draw_networkx(G)
```

We can also add nodes along with node attributes if our container yields 2-tuples of the form

 (node, node_attribute_dict)

```python
# os atributos color apenas identificam o nó, mas não afetam sua aparência visual
G.add_nodes_from([(5,{"color":"red"}),(4,{"color":"green"})])
G.nodes.data()
```

```python
nx.draw_networkx(G)
```

Nodes from one graph can be incorporated into another.

```python
T = nx.path_graph(15)
nx.draw_networkx(T)
```

```python
G.add_nodes_from(T)
nx.draw_networkx(G)
```

We can have Graph attributes, Node attributes and Edge attributes.

####Graph attributes

```python
G =  nx.Graph(day='Friday')
# We can modify attributes later
G.graph['day'] = 'Tuesday'
G.graph
```

####Node attributes

```python
G.add_node(1,time='5pm')
G.add_nodes_from([3],time='1pm')
G.nodes[1]
# We can add several attributes for each graph/node/edge
G.nodes[1]['room'] = 812
G.nodes.data()
```

##Edges
G can also be grown by adding one edge at a time.

```python
G = nx.Graph()
G.add_edge(1,2)
e = (2,3)
G.add_edge(*e) # unpack edge tuple*

T = nx.path_graph(20)
G.add_edges_from(T.edges)
nx.draw_networkx(G)
```

We can empty our graph.

```python
G.clear()
```

There are no complaints when adding existing nodes or edges.

If we add new nodes/edges and NetworkX quietly ignores any that are already present.

```python
G.add_edges_from([(1,2),(1,3)])
nx.draw_networkx(G)
```

```python
G.add_edge(1,2)
nx.draw_networkx(G)
```

```python
# Adioionaer um nó por nome ou nós a partir de de uma string
# add nó por nome
G.add_node("paper")
# add nó pela string
G.add_nodes_from("paper")
nx.draw_networkx(G)
```

```python
# adicione uma aresta (vínculo) entre os nós "p" e "3"
G.add_edge(3,'p')
nx.draw_networkx(G)
```

```python
# mostra o número de nós e arestas do grafo
G.number_of_nodes(), G.number_of_edges()
```

#### Atributos das arestas (edges)

```python
G = nx.Graph()
G.add_edge(1,2,weight=5.3)
G.add_edges_from([(3,4),(4,5)],color='green')
G.add_edges_from([(1,2,{'color':'blue'}),(2,3,{'weight':8})])
G.edges[2,3]['color'] = 'red'
G.edges[3,4]['weight'] = 6.7
G.edges[4,5]['weight'] = 11
G.edges.data()
```
