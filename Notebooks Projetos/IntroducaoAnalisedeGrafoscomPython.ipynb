#  Introdução a Análise de Redes Sociais (ARS)

A análise de redes sociais é uma disciplina interdisciplinar que investiga a estrutura, dinâmica e comportamento das redes sociais. As redes sociais não se limitam apenas às plataformas online, mas abrangem interações entre indivíduos, organizações ou entidades em diversos contextos, como amizades, colaborações profissionais, trocas de informações e influências sociais (LEMEUX;OUIMET, 2008).

O conceito de Rede Social surgiu há séculos, nas áreas de antropologia social e sociologia, porém, apenas nas últimas décdas, os estudos em redes sociais vem se intesificando, com a possibilidade de anlisar as características individuais de seus componentes, como também estudadr as características estruturais da rede (DIGIAMPIETRI, 2023).

A análise de redes sociais está intrinsecamente ligada à teoria dos grafos, uma vez que ambas compartilham o conceito fundamental de representar interações entre entidades por meio de nós e arestas. Na teoria dos grafos, os nós representam entidades individuais, como pessoas ou organizações, enquanto as arestas denotam as conexões ou relações entre essas entidades. Da mesma forma, na análise de redes sociais, os nós representam indivíduos ou entidades sociais, enquanto as arestas refletem os laços sociais, como amizades, colaborações ou trocas de informações.

## Grafos

Grafos são estruturas matemáticas (ou modelos matemáticos) que permitem codificar relacionamentos entre pares de obejtos.

Os objetos são:

* Os nós ou vértices dos grafos;
* As reçaões entre os nós são as arestas.

Grafos podem ser utilizaod para representar relações de forma visual, usando pontos ou círculos para represenatr os nós e linhas para representar as arestas.

Grafos podem ser usados para modelar diversos fenômenos da realidade, como redes sociais, mapas, circuitos, algoritmos, entre outros.

### Nós

Representam entidades, sejam elas pessoas, perfis do Twitter, empresas, páginas ou usuários do Facebook, canais ou vídeos do Youtube, etc. Estes nós podem possuir atributos ou características específicas, tais como tipos, tamanho e peso.



### Arestas

São as conexões entre os nós, representando as ligações entre as entidades.

Em uma rede social, por exemplo, estas relações podem representar interações como curtidas, retweets (no caso do Twitter/X) ou conexões entre canais de YouTube.



### Tipos de grafo

#### Grafo não direcionado

Um grafo onde as relações entre as arestas não têm sentido definido, representando uma relação bidirecional entre os vértices.

Um exemplo de grafo não direcionado seria uma relação de amizade em uma rede social, o vínculo (aresta) estabelece a relação entre duas pessoas sem usar um sentido.

No exemplo a seguir, Carlos é amigo de Maria e João.

<p align="center"><img src="https://github.com/paulosrl/imagens/blob/master/redenaodirecional.png?raw=true"></p>

#### Grafo direcionado

Um grafo no qual as relações represntadas pelas arestas têm sentido definido, indicando clramente a relação de um nó origem para um nó destino.

As as arestas dvem ser percorridas em uma direção única. 

Este tipo da gráfico admite que um nó se relacione com ele mesmo (self-loop). Vale resssltar que self-loops não são permitidos em grafos não direcionados.

A ligação entre os nós não é necessariamente mútua, por exemplo:

No Instagram, por exemplo, é possível seguir um perfil sem que este perfil siga quem o está seguindo.

No exemplo a seguir João e Maria seguem Carlos, mas Carlos não segue João e Maria.

# Criando grafos direcionados


```python
# cria um grafo direcionado, o sentido da seta é sempre do nó da esquerda para direita
G = nx.DiGraph([(0,3),(1,3),(2,4),(3,6),(5,6)])
nx.draw_networkx(G)
```

```python
# cria grupo de nós por colunas
nos_esquerda = [0,1,2]
nos_meio = [3,4]
nos_direita = [5,6]
# ajustando a posição dos nós de acordo com as colunas
# posiciona os nós de acordo com o grupo
# nós a esquerda
pos = {n: (0,i) for i,n in enumerate(nos_esquerda)}
# nós do meio
pos.update({n: (1,i+0.5) for i,n in enumerate(nos_meio)}) # type: ignore
# nós a direita
pos.update({n: (2,i+0.5) for i,n in enumerate(nos_direita)}) # type: ignore
# desenha a rede com o atributo pos
nx.draw_networkx(G,pos)
```

## Formato Kamada Kawai

```python
G = nx.path_graph(8)
pos = nx.kamada_kawai_layout(G)
nx.draw_networkx(G,pos)
```

## Formato planar

```python
G = nx.path_graph(8)
pos = nx.planar_layout(G)
nx.draw_networkx(G,pos)
```

## Formato randômico

```python
G = nx.lollipop_graph(4, 3)
pos = nx.random_layout(G)
nx.draw_networkx(G,pos)
```

## Formato shell

```python
G = nx.path_graph(8)
shells = [[0], [1, 2, 3,4,5,6,7]]
pos = nx.shell_layout(G, shells)
nx.draw_networkx(G,pos)
```

## Formato spring

```python
G = nx.path_graph(8)
pos = nx.spring_layout(G)
nx.draw_networkx(G,pos)
```

## Formato spectral

```python
G = nx.path_graph(8)
pos = nx.spectral_layout(G)
nx.draw_networkx(G,pos)
```

## Formato Espiral

```python
G = nx.path_graph(8)
pos = nx.spiral_layout(G)
nx.draw(G, pos=pos)
nx.draw_networkx(G,pos)
```

## Formato bipartido

Um grafo bipartido é um tipo de grafo que tem seus vértices divididos em dois conjuntos disjuntos, de tal forma que não há arestas entre vértices do mesmo conjunto.

```python
G = nx.bipartite.gnmk_random_graph(3, 5, 10, seed=123)
top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G, top)
nx.draw_networkx(G,pos)
```

## Formato multipartido

```python
# Um grafo multipartido é um tipo de grafo que tem seus vértices
# divididos em vários conjuntos disjuntos, de tal forma que não
# há arestas entre vértices do mesmo conjunto.
G = nx.complete_multipartite_graph(10, 5, 3)
pos = nx.multipartite_layout(G)
nx.draw_networkx(G,pos)
```
