##Import of Libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import approximation as approx
```

## Clique

```python
G = nx.Graph()
G.add_edges_from([(1,2),(2,3),(3,4),(1,5),(2,5),(3,5)])

nx.draw(G,with_labels=True)

# Find the maximum clique

clique = approx.max_clique(G)

# Print the maximum clique

print("Maximum Clique:")
print(clique)
```

## Maximum Independent Set

```python
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5),(5,6)])


nx.draw(G,with_labels=True)

# Find the maximum independent set
independent_set = approx.maximum_independent_set(G)

# Print the maximum independent set
print("Maximum Independent Set:")
print(independent_set)
```
