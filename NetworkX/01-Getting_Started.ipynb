##Import of Libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
```

##Simple Graph
A graph consists of nodes (also called vertices) that are connected by edges.

```python
#instancia um novo grafo (G)
G = nx.Graph()

#define as ligações entre arestas
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(1,5)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)
G.add_edge(5, 6)

# posição dos nós na rede
pos = {1:(0,0),2:(-1,0.35),3:(2,0.18),4:(4,0.255),5:(5,0.06),6:(6,0.300)}
# aparência dos nós
# linewidths: epessura do contorno dos nós 
# width: espessura das arestas
options = {"font_size": 20,"node_size":2000,"node_color":"yellow",
           "edgecolors":"black","linewidths":3,"width":5}

# constrói a rede usando os parâmetros de posição (pos) e layout(options)
#nx.draw_networkx(G,pos,**options)

nx.draw_networkx(G,**options)

# constrói a rede de forma aleatória
#nx.draw_networkx(G,pos)

# (**) takes a dictionary and extends it into a series of arguments

# Set margins for the axes so nodes aren't clipped
ax = plt.gca()

#define uma margem na figura
#ax.margins(0.20)

# desabilta moldura
plt.axis("off")
# mostra a rede
plt.show()
```
