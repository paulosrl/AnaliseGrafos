##Import of Libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Average Clustering

```python
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (3, 5)])

nx.draw(G,with_labels=True)

#
```

```python
### cáluclo do coef. de clustering da rede (médio)

avg_clustering = nx.average_clustering(G)

print("Média do coeficiente de clustering : ",avg_clustering)#
```
