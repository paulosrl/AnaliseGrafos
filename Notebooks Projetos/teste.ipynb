```python
import pandas as pd
import networkx as nx

# Ler o arquivo CSV
df = pd.read_csv('nosarestas.csv')

# Criar um grafo vazio
G = nx.Graph()

# Adicionar arestas ao grafo
for _, row in df.iterrows():
    G.add_edge(row['origem'], row['destino'], weight=row['peso'])

# Agora G é um grafo com arestas e pesos

pos = nx.spring_layout(G)  # Definir a posição dos nós
nx.draw(G, pos, with_labels=True)  # Desenhar nós e arestas
edge_labels = nx.get_edge_attributes(G, 'weight')  # Obter os pesos das arestas
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels);  # Adicionar rótulos às arestas
```

```python
# Definir a posição dos nós
pos = nx.spring_layout(G)

# Obter os pesos das arestas
weights = nx.get_edge_attributes(G, 'weight')

# Normalizar os pesos para o intervalo [0, 1] para usar como espessura da linha
max_weight = max(weights.values())
normalized_weights = [2 * w / max_weight for w in weights.values()]  # Multiplicar por 5 para aumentar a espessura

# Desenhar nós e arestas
nx.draw(G, pos, with_labels=True, edge_color='black', width=normalized_weights)  # Arestas vermelhas
```
